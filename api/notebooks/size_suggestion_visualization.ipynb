```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_test.csv')
```

```python
df.duplicated().sum()
```

```python
size_order = ['XXS', 'XS', 'S', 'M', 'L', 'XL', 'XXL', 'XXXL']

fig, axes = plt.subplots(2, 4, figsize=(20, 10), sharey=True)
axes = axes.flatten() 

for i, size_label in enumerate(size_order):
    size_df = df[df['size'] == size_label]
    
    if not size_df.empty:
        m_averages = size_df[['age', 'height', 'weight']].mean().reset_index()
        m_averages.columns = ['Metric', 'Avg']
        
        sns.barplot(data=m_averages, x='Metric', y='Avg', ax=axes[i], palette='magma')
        
        for p in axes[i].patches:
            axes[i].annotate(format(p.get_height(), '.1f'), 
                             (p.get_x() + p.get_width() / 2., p.get_height()), 
                             ha='center', va='center', xytext=(0, 7), 
                             textcoords='offset points', fontsize=10)
        
        axes[i].set_title(f'Size: {size_label}', fontsize=14, fontweight='bold')
        axes[i].set_xlabel('')
        axes[i].set_ylabel('Average Value' if i % 4 == 0 else '')
    else:
        axes[i].set_title(f'Size: {size_label} (No Data)')

plt.tight_layout()
plt.suptitle('Physical Metrics Comparison Across All Sizes', fontsize=20, y=1.05)
plt.show()
```

```python
size_order = ['XXS', 'XS', 'S', 'M', 'L', 'XL', 'XXL', 'XXXL']

# 2. Setup the figure (2 rows, 4 columns for 8 sizes)
fig, axes = plt.subplots(2, 4, figsize=(20, 10), sharey=True)
axes = axes.flatten() # Flatten the 2D array of axes for easy looping

# 3. Loop through each size and create a bar chart
for i, size_label in enumerate(size_order):
    # Filter data for the specific size
    size_df = df[df['size'] == size_label]
    
    if not size_df.empty:
        # Calculate averages for Age, Height, Weight
        m_averages = size_df[['age', 'height', 'weight']].mean().reset_index()
        m_averages.columns = ['Metric', 'Avg']
        
        # Plot on the specific subplot axis
        sns.barplot(data=m_averages, x='Metric', y='Avg', ax=axes[i], palette='magma')
        
        # Add values on top of bars
        for p in axes[i].patches:
            axes[i].annotate(format(p.get_height(), '.1f'), 
                             (p.get_x() + p.get_width() / 2., p.get_height()), 
                             ha='center', va='center', xytext=(0, 7), 
                             textcoords='offset points', fontsize=10)
        
        axes[i].set_title(f'Size: {size_label}', fontsize=14, fontweight='bold')
        axes[i].set_xlabel('')
        axes[i].set_ylabel('Average Value' if i % 4 == 0 else '')
    else:
        axes[i].set_title(f'Size: {size_label} (No Data)')

# 4. Adjust layout so titles don't overlap
plt.tight_layout()
plt.suptitle('Physical Metrics Comparison Across All Sizes', fontsize=20, y=1.05)
plt.show()
```
