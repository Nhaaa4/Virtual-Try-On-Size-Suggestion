```python
import os
import shutil
import random
from pathlib import Path
from collections import defaultdict

DATA_ROOT   = Path(r"d:\CADT\Year_3-Term_2\Capstone\image-classification\DATA")
SPLITS      = {"train": 0.80, "validation": 0.10, "test": 0.10}
IMG_EXTS    = {".jpg"}
source_classes = ["bikini", "bra", "underwear"]   
```

```python
summary = defaultdict(dict)

for cls_name in source_classes:
    cls_dir = DATA_ROOT / cls_name
    images = [f for f in cls_dir.iterdir() if f.suffix.lower() in IMG_EXTS]

    if not images:
        print(f"  [{cls_name}] No images found, skipping.")
        continue

    random.shuffle(images)
    n = len(images)
    n_train = int(n * SPLITS["train"])
    n_val   = int(n * SPLITS["validation"])

    split_map = {
        "train":      images[:n_train],
        "validation": images[n_train : n_train + n_val],
        "test":       images[n_train + n_val :],
    }

    for split_name, files in split_map.items():
        dest_dir = DATA_ROOT / split_name / cls_name
        dest_dir.mkdir(parents=True, exist_ok=True)
        for f in files:
            shutil.copy2(f, dest_dir / f.name)
        summary[cls_name][split_name] = len(files)

    print(f"[{cls_name}]  total={n}  "
          f"train={summary[cls_name]['train']}  "
          f"val={summary[cls_name]['validation']}  "
          f"test={summary[cls_name]['test']}")
```

```python
print("── Final counts in DATA/ ──")
for split in ["train", "validation", "test"]:
    split_dir = DATA_ROOT / split
    if not split_dir.exists():
        continue
    print(f"\n{split}/")
    for cls_dir in sorted(split_dir.iterdir()):
        imgs = [f for f in cls_dir.iterdir() if f.suffix.lower() in IMG_EXTS]
        print(f"  {cls_dir.name:20s}: {len(imgs):>5} images")
```
